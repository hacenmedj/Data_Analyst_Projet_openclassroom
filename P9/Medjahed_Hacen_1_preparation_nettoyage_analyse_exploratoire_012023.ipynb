# Préparation, nettoyage et l’analyse exploratoire des données

## Importation des données

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DispoA=pd.read_csv("C:/Users/hotsa/Downloads/projetDATANalyst/p9/DAN-P9-data/DisponibiliteAlimentaire_2017.csv")
Popu=pd.read_csv("C:/Users/hotsa/Downloads/projetDATANalyst/p9/DAN-P9-data/Population_2000_2018.csv")
```

## Nettoyage des données

```python
## choisir que les données du ficher  population pour l'année 2017
Popu2017=Popu[Popu['Année']==2017]
Popu2017=Popu2017[['Zone','Valeur']]
Popu2017.columns=['Zone','Population']
Popu2017.head()
```

```python
Popu2017.columns
```

```python
## avoir une vue des données du ficher Disponibilité Alimentaire
DispoA.head()
```

```python
## avoir les informations du ficher disponobilité alimentaire
DispoA.info()
```

```python
## garder que les lignes avec le produit = "viande de volaille"
ViandeDV=DispoA[DispoA.Produit.str.contains("Volaille")]
ViandeDV
```

```python
## selection des colonnes qui nous interesse pour notre étude
ViandeDV=ViandeDV[["Élément","Zone","Valeur"]]
ViandeDV
```

```python
## Réorganisation des dataframes grace à pivot
## les donnees de zone en ligne et les données d'élément en colonnes
ViandeDV=ViandeDV.pivot(index="Zone",columns="Élément",values="Valeur")
ViandeDV
```

```python
## on selectionne les colonnes qui nous interesse:
ViandeDV=ViandeDV[["Exportations - Quantité","Importations - Quantité","Production","Disponibilité intérieure","Disponibilité alimentaire en quantité (kg/personne/an)"]]
ViandeDV
```

```python
## suppression de toute les valeures manquantes
ViandeDV=ViandeDV.fillna(value=0)
ViandeDV.info()
```

```python
## faire le merge des données entre popu2017 et viandeDV gràce à la colonne Zone
ViandeDV=pd.merge(Popu2017,ViandeDV,on='Zone')
## permet de voir ce que les premieres lignes de ce que l'on a obtenus
ViandeDV.head()
```

```python
## vérification si il y des lignes qui se repete
ViandeDV.duplicated()
```

```python
## verifier si il y a des valeures manquantes:
ViandeDV.isnull().sum()
```
